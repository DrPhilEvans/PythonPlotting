```python
# Import the modules we need, or may need. Include lots of possible formatters so we have options.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
from matplotlib.ticker import LogLocator, LogFormatterSciNotation, LogFormatter, FuncFormatter
from matplotlib.colors import LogNorm, Normalize, CenteredNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import math
```

## Load the data from the file.

Your data file needs at least 3 columns: x, y and the property to be plotted.

To load the data, edit the next cell as follows:
* Replace "demoData/surface-ripple.dat" with your data file.
* If you want to, replace: `X, Y, data` with variable names of your choice.
* If you have data in more than one file, you will need more than one `loadContourData` command -- with its own filenames and variables. I have given one example (commented out).

```python
# This loads the data, assuming your file has columns: x, y, dx, dy
# where (x,y) is the position of the arrow, dx is the magnitude in the
# x-direction, and dy is the magnitude in the y-direction.
xv, yv, dx, dy = np.loadtxt("demoData/some-arrows.dat", unpack=True)
```

## Plot the data

Edit this cell to plot the data you want to plot. First:
* Change the `ax.quiver` line, replacing the "xv, yv, dx, dy"" with the variables use loaded in the previous cell.

There are empty sections, marked with comments (starting `#`). These are places where you can insert
lines from the [Other Plot Controls](other_plot_controls.ipynb) notebook to customise the appearance of the plot.

That notebook also shows you how to customise the `ax.quiver` command.

I recommend running the cell first without adding any controls in, so you can see what the data look like, and then copy lines in as needed.

```python
fig, ax = plt.subplots(figsize=(6, 6))  # Size is in inches(!)

# Set axis labels

# Set tick label size

# Format tick labels

# Plot the arrows
ax.quiver(
    xv, yv, dx, dy,
    scale=40,          # smaller = longer arrows
    width=0.005,       # line thickness
    headwidth=10,      # Width of arrow head
    headlength=15,     # Length of arrow head from rear 'points' to tip
    headaxislength=10, # Length of arrow head from shaft to tip.
    color="blue"
)

# Prevent axis labels falling outside the figure.
fig.tight_layout()
```

## Save the figure

When you are happy with your figure:

**Note**: For surface plots, the image is always a bitmap, even in pdf or svg form. We can force it to be high quality (so that scaling should not hurt us much) by specifying the resolution in dots per inch (dpi) that it should be written at.

```python
fig.savefig("vector.svg", facecolor='white', transparent='False', dpi=300)
```
