```python
# Import the modules we need, or may need. Include lots of possible formatters so we have options.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
from matplotlib.ticker import LogLocator, LogFormatterSciNotation, LogFormatter, FuncFormatter
from matplotlib.colors import LogNorm, Normalize, CenteredNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import math

# Also define a function to load the data
def loadImageData(filename):
    with open(filename) as f:
        xlo, xhi, ylo, yhi = map(float, f.readline().split())
        data = np.loadtxt(f)
        data = data[:-1, :-1]  # Need to chop off the last
        return xlo, xhi, ylo, yhi, data
```

## Load the data from the file.

Your data file needs at least 3 columns: x, y and the property to be plotted. To make a surface plot we need to do some extra work reading the file, so I have provided a special function `loadImageData()` in the cell above, which you call here. 

To load the data, edit the next cell as follows:
* Replace "demoData/surface-ripple.dat" with your data file.
* If you want to, replace: `xlo, xhi, ylo, yhi, data` with variable names of your choice.


```python
xlo, xhi, ylo, yhi, data = loadImageData("demoData/surface-ripple.dat")
```

## Plot the data

Edit this cell to plot the data you want to plot. First:
* Change the `ax.imshow` line replacing "data", "xlo", "xhi", "ylo", "yhi" with the variables you set in the previous cell.

There are empty sections, marked with comments (starting `#`). These are places where you can insert
lines from the [Other Plot Controls](other_plot_controls.ipynb) notebook to customise the appearance of the plot.

That notebook also shows you how to customise the `ax.imshow` command, and the colourbar.

I recommend running the cell first without adding any controls in, so you can see what the data look like,
and then copy lines in as needed.

```python
fig, ax = plt.subplots(figsize=(6, 6))  # Size is in inches(!)

# Set axis labels

# Set tick label size

# Format tick labels

# Plot the data
sc = ax.imshow(data, origin="lower", extent=(xlo, xhi, ylo, yhi), aspect=1, cmap="cool")

# Draw a colourbar
cbar = fig.colorbar(sc, ax=ax, format="%.2f", location="right", shrink=0.73)
cbar.set_label(label="Temperature (K)", fontsize=16)

# Prevent axis labels falling outside the figure.
fig.tight_layout()
```
